## Feature design



- Choix des attributs

> Afin de choisir quels attributs du dataset sont pertinent pour la detection d'url de phishing, nous avons calculer la matrice de covariance des attributs du set

> De plus afin de favoriser la précision du modèle nous avons, après avoir éliminé les attributs non numérique, normalisé entre 0 et 1 les attributs restants

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from datetime import datetime
from sklearn.datasets import fetch_openml
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import normalize
# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)

def filter(L):
    res = []
    for i in L:
        if i > 0.5:
            res.append(1)
        else:
            res.append(0)
    return res


def compute_accuracy(Y_true, Y_pred):
    correctly_predicted = 0
    # iterating over every label and checking it with the true sample
    for true_label, predicted in zip(Y_true, Y_pred):
        if true_label == predicted:
            correctly_predicted += 1
    # computing the accuracy score
    accuracy_score = correctly_predicted / len(Y_true)
    return accuracy_score


def build_set(path):
    df = pd.read_csv(path, header=0)
    x = df._get_numeric_data()
    x = x.drop(columns=['Unnamed: 0'])
    x = x.to_numpy()
    x = normalize(x, axis=0)
    y = df["status"].map({'legitimate': 0, 'phishing': 1})
    corr_matrix = df.corr()
    sns.heatmap(corr_matrix, annot=True)
    plt.show()
    #print(corr_matrix)
    
    return x, y
```

```python
# Define Grid
x_train, y_train = build_set("train.csv")
x_train, x_test, y_train, y_test = train_test_split(
    x_train, y_train, test_size=0.3, shuffle=True)

```

## Méthodologie

- Répartition des données

> Nous avons réalisé un split avec 70% des données pour l'entrainement et 30% pour tester le modèle, ce split permet de detecter un overfit du modèle

> Le dataset contient 4629 url legitime et 3543 url compromises, on peut dire que le dataset est plustot équilibré, nous avons donc choisi de ne pas reéquilibré ce dernier

- Choix du modèle

> Après lecture d'un article nous avons choisi d'utiliser un random forest classieur, étant un modèle délivrant le plus de résultats pour ce genre d'exercice

- Tuning des hyperparamètres


    Afin de déterminer les hyperparamètre optimaux nous effectuons un grid search sur les paramètres suivant:

        - Le nombre d'estimateurs
        - La fonction pour le calcul du nombre de features à regarder
        - et la profondeur de l'arbre

```python
grid = {
    'n_estimators': [1000, 1200, 1400, 1600],
    'max_features': ['sqrt','log2'],
    'max_depth': [24, 28, 30, 32],
    'random_state': [18]
}
```

```python
CV_rfr = GridSearchCV(estimator=RandomForestClassifier(),
                      param_grid=grid, cv=5, verbose=10, n_jobs=-1)
CV_rfr.fit(x_train, y_train)
```

Les hyper paramètres optimaux trouvés par le grid search sont les suivants:

```python
print("\n The best parameters across ALL searched params:\n", CV_rfr.best_params_)
```

On peut maintenant prédire sur le jeux de test pour obtenir la précision du modéle et la matrice de confusion

```python
prediction = filter(CV_rfr.predict(x_test))
```

```python
print("Accuracy: ", compute_accuracy(y_test, prediction))
```
